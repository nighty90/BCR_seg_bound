```python
from Bio import SeqIO
import pandas as pd
```

```python
DATA_PATH = "./data/implant_shm_seed24_100_000.fasta"
df = pd.DataFrame()
for i, seq_record in enumerate(SeqIO.parse(DATA_PATH, "fasta")):
    df.loc[i, "sequence"] = str(seq_record.seq)
    id_elements = seq_record.id.split(":")
    id_fields = [
        "v_call", "d_call", "j_call", 
        "V3P_length", "D5P_length", "D3P_length", "J5P_length",
        "V3D_length", "D5D_length", "D3D_length", "J5D_length",
        "np1_length", "np2_length", "np1", "np2",
        "np1_sequence_start", "d_sequence_start", "np2_sequence_start", "j_sequence_start",
        "productivity", "index", "evolution_time"
    ]
    df.loc[i, id_fields] = id_elements[:-1]
df.head()
```

```python
seq = df.loc[0, "sequence"]
np1_start = int(df.loc[0, "np1_sequence_start"])
np1_end = int(df.loc[0, "d_sequence_start"])
np2_start = int(df.loc[0, "np2_sequence_start"])
np2_end = int(df.loc[0, "j_sequence_start"])
print(seq[np1_start:np1_end])
print(seq[np2_start:np2_end])
```

```python
df_new = pd.DataFrame()
df_new["sequence"] = df["sequence"]
df_new["v_sequence_start"] = 1
df_new["v_sequence_end"] = df["np1_sequence_start"].astype("int32")
df_new["d_sequence_start"] = df["d_sequence_start"].astype("int32") + 1
df_new["d_sequence_end"] = df["np2_sequence_start"].astype("int32")
df_new["j_sequence_start"] = df["j_sequence_start"].astype("int32") + 1
df_new["j_sequence_end"] = df["sequence"].str.len()
df_new.head()
```

```python
df_new.to_csv("./data/implant_shm_seed24_100_000.tsv", sep="\t")
```
